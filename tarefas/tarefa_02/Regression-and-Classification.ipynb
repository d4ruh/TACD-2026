### Degree vs. RMSE
Generate the plot (as in __Figure 1__) of the RMSE during training and testing at different values of of the polynomial degree $M$.

```python
# your code goes here:
def basis_func_generator(degree=1):
    def basis_func(X):
        return np.column_stack([X ** i for i in range(degree + 1)])
    
    return basis_func

def polynomial_regression(X, y, basis_func=basis_func_generator(1)):
    phi = basis_func(X)
    pinv = np.linalg.inv(phi.T @ phi) @ phi.T 
    return pinv @ y

def RMSE(y_true, y_predict):
    return ( (y_true - y_predict) ** 2 ).mean() ** (1/2)

funcs = [basis_func_generator(i) for i in range(10)]
weights_by_degree = [polynomial_regression(x_train, y_train, f) for f in funcs]

# # testing how the weights behave:
# sns.scatterplot(x=x_train, y=y_train, label='Train Points', color='blue')
# sns.scatterplot(x=x_test, y=y_test, label='Test Points', color='black')
# sns.lineplot(x=np.linspace(0, 10, num=200), 
#              y=funcs[9](np.linspace(0, 10, num=200)) @ polynomial_regression(x_train, y_train, funcs[9]),
#              label='Prediction', color='red')
# sns.lineplot(x=np.linspace(0, 10, num=200), 
#              y=4 * np.sin(np.linspace(0, 10, num=200)),
#              label='4 * sin(x)', color='green', linestyle='--', alpha=0.5)
# plt.show()

resultados = []
for i in range(10):
    data = {'degree': i,
            'dataset':'training',
            'RMSE': RMSE(y_train, funcs[i](x_train) @ weights_by_degree[i])}
    resultados.append(data)

    data = {'degree': i,
            'dataset':'testing',
            'RMSE': RMSE(y_test, funcs[i](x_test) @ weights_by_degree[i])}
    resultados.append(data)

resultados_df = pd.DataFrame(resultados)
sns.lineplot(resultados_df, x='degree', y='RMSE', hue='dataset', style='dataset', markers=True)
plt.grid(True)
plt.show()
```

### Training set size vs. RMSE
Generate the plot (as in __Figure 2__) of the RMSE during training and testing at different training set sizes $N$.

```python
# your code goes here:

basis_func = basis_func_generator(9)
resultados = []
for i in range(50):
    n = 10 * (i + 1)
    x = np.linspace(0, 10, num=n)
    noise = rng.normal(0, 1, size=n)
    y = 4 * np.sin(x) + noise

    weights = polynomial_regression(x, y, basis_func)

    data = {
        'training_size': n,
        'dataset': 'training',
        'RMSE': RMSE(y, basis_func(x) @ weights)
    }
    resultados.append(data)

    data = {
        'training_size': n,
        'dataset': 'testing',
        'RMSE': RMSE(y_test, basis_func(x_test) @ weights)
    }
    resultados.append(data)

resultados_df = pd.DataFrame(resultados)
sns.lineplot(resultados_df, x='training_size', y='RMSE', hue='dataset')
plt.grid(True)
plt.show()
```

### Here you will implement the KNN algorithm using the euclidean distance

```python
def euclidean_dist(X, x_predict): 
    # your code goes here:
    return np.sum((X - x_predict) ** 2, axis=1) ** (1/2)
    
```

```python
def knn_predict(k, X, y, x_pred):
    # your code goes here:
    dist = euclidean_dist(X, x_pred)
    sorted_idx = np.argsort(dist)
    knn = sorted_idx[:k]

    classes, freq = np.unique(y[knn], return_counts=True)
    return classes[freq.argmax()]

def knn_predict_batch(k, X, y, X_predict):
    return np.array([knn_predict(k, X, y, x_i) for x_i in X_predict])
```

### Select the optimal $K$ by plotting the F1 score in the validation set at different values of $K$

First create a function that calculates the F1-score

```python
def calculate_f1_score(y_true, y_predict):
    # your code goes here:
    np_y_true = np.array(y_true)
    np_y_predict = np.array(y_predict)

    conf_mtx = [[0, 0], [0, 0]]

    conf_mtx[0][0] = ((np_y_true == 1) & (np_y_predict == 1)).astype(int).sum() # True Positive
    conf_mtx[0][1] = ((np_y_true == 0) & (np_y_predict == 1)).astype(int).sum() # False Positive
    conf_mtx[1][0] = ((np_y_true == 1) & (np_y_predict == 0)).astype(int).sum() # False Negative
    conf_mtx[1][1] = ((np_y_true == 0) & (np_y_predict == 0)).astype(int).sum() # True Negative

    # f1_score = TP / (TP + (FN + FP) / 2)
    return conf_mtx[0][0] / (conf_mtx[0][0] + (conf_mtx[1][0] + conf_mtx[0][1]) / 2 )

y_io_validation_predict = knn_predict_batch(5, x_io_train, y_io_train, x_io_validation)
print(y_io_validation_predict == y_io_validation)
print(calculate_f1_score(y_io_validation, y_io_validation_predict))
    
```

Generate the plot and select the optimal $K$

```python
# your code goes here:

knn_scores = [{
    'K': i,
    'F1-score': calculate_f1_score(y_io_validation, knn_predict_batch(i, x_io_train, y_io_train, x_io_validation)),
    'dataset': 'validation'
}
for i in range(1, 32, 2)]

knn_scores_df = pd.DataFrame(knn_scores)

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.lineplot(data=knn_scores_df, x='K', y='F1-score', hue='dataset', style='dataset', markers=True, ax=ax)
ax.grid(True)
ax.set_xticks(knn_scores_df['K'])
plt.show()
```

Calculate the test set performance using the optimal $K$

```python
# your code goes here:
best_k = knn_scores_df.loc[knn_scores_df['F1-score'].argmax()]['K']

x_io_merged = np.vstack((x_io_train, x_io_validation))
y_io_merged = np.vstack((y_io_train.reshape((-1, 1)), y_io_validation.reshape((-1, 1))))
f1_score = calculate_f1_score(y_io_test, knn_predict_batch(best_k, x_io_merged, y_io_merged, x_io_test))

print('F1-score para o melhor K no conjunto de teste: ', f1_score)
```
