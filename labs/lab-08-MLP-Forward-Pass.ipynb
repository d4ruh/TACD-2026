Now you will write the function that computes the forward pass. 
* I provide here a structure that you can follow for your function, but again, feel free to modify it as you see fit.
* Use the sigmoid function as the activation of the hidden layer.
* Don't forget about the biases!
* *It is up to you to think what should be the activation for the output layer.*

```python
def sigmoid_activation(z):
    # your code goes here
    return 1/(1+np.exp(-z))
```

Para cada entrada $i$ do nosso vetor $X$ temos, para $j = 0, 1, ..., dim\_hidden$ no primeiro nível, o seguinte cálculo:
$$
a_{ij} = b_{j} + \sum_{k=1}^{dim\_input}{W^{(1)}_{kj}X_{ik}}
$$
E para uma função de ativação $f$:
$$
z_{ij} = f(a_{ij})
$$

Logo:
$$
a_i = b^T + X_i W^{(1)}
$$

Com isso temos as entradas para o segundo nível, onde repetimos o processo alterando a dimensionalidade


```python
def two_layer_perceptron(X, activation, dim_input, dim_hidden, dim_output):
    """
    Implements the forward pass of a two-layer fully connected perceptron.
    
    Parameters
    ----------
    X : a 2-dimensional array
        the input data
    activation : function
        the activation function to be used for the hidden layer
    dim_input : int
        the dimensionality of the input layer
    dim_hidden : int
        the dimensionality of the hidden layer
    dim_output : int
        the dimensionality of the output layer
    Returns
    -------
    y_pred : float
        the output of the computation of the forward pass of the network
    """
    # your code goes here

    # pesos do primeiro nível:
    W1 = np.random.normal(0, 1, (dim_input, dim_hidden)) 
    b1 = np.random.normal(0, 1, dim_hidden)
    A1 = b1 + X @ W1
    Z1 = activation(A1)

    # pesos do segundo nível:
    W2 = np.random.normal(0, 1, (dim_hidden, dim_output))
    b2 = np.random.normal(0, 1, dim_output)
    return b2 + Z1 @ W2
    
```

Calculate the RMSE of the forward pass. 

```python
# your code goes here

def RMSE(y_true, y_pred):
    return (np.mean((y_true - y_pred) ** 2)) ** (1/2)

y_pred = two_layer_perceptron(x_train, sigmoid_activation, x_train.shape[1], 5, 1)
print(RMSE(y_train, y_pred))
```

Write a function that calculates the forward pass for this multi-class classification problem.
* You will use the sigmoid activation function for the hidden layer.
* For the output layer you will have to write the softmax activation function (you can check the slides)
* __Note:__ you can easily re-use the function that you coded for Part I if you do a simple modification and also include an input argument for the activation of the output layer.

```python
def softmax_activation(z):
    # your code goes here:
    exp = np.exp(z)
    return exp / np.sum(exp, axis=1)[:, None]
```

```python
# your code goes here: 
y_pred = two_layer_perceptron(x_train, sigmoid_activation, x_train.shape[1], 10, 10)
y_pred = softmax_activation(y_pred)
```

Lastly, calculate the error of this forward pass using the cross-entropy loss.

```python
# your code goes here:
def multiclass_cross_entropy(y_true, y_pred):
    # dado que os valores de y_true representam a classe (de 0 a 9) que um datapoint ocupa,
    # podemos calcular a função de erro como (-1) x a soma dos logaritmos da previsão da prob
    # da classe correta dos datapoints

    erro = 0
    for t, pred in zip(y_true, y_pred):
        erro -= np.log(pred[t])

    return erro

print(multiclass_cross_entropy(y_train, y_pred))
```
